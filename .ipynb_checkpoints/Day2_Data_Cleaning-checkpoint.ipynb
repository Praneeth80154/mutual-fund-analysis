```python
import pandas as pd

nav = pd.read_csv("data/raw/02_nav_history.csv")

print("Shape:", nav.shape)

print("\nMissing Values:")
print(nav.isnull().sum())

print("\nDuplicate Rows:", nav.duplicated().sum())

print("\nNAV <= 0:")
print((nav["nav"] <= 0).sum())
```

```python
nav["date"] = pd.to_datetime(nav["date"])

nav = nav.sort_values(
    by=["amfi_code", "date"]
)

nav = nav.drop_duplicates()

nav["nav"] = nav.groupby("amfi_code")["nav"].ffill()

nav = nav[nav["nav"] > 0]

print("Cleaned Shape:", nav.shape)
```

```python
print("Missing Values:")
print(nav.isnull().sum())

print("\nDuplicate Rows:")
print(nav.duplicated().sum())

print("\nNAV <= 0:")
print((nav["nav"] <= 0).sum())
```

```python
print("Shape:", transactions.shape)

print("\nMissing Values:")
print(transactions.isnull().sum())

print("\nDuplicate Rows:")
print(transactions.duplicated().sum())

print("\nAmount <= 0:")
print((transactions["amount_inr"] <= 0).sum())

print("\nKYC Values:")
print(transactions["kyc_status"].unique())
```

```python
transactions["transaction_date"] = pd.to_datetime(
    transactions["transaction_date"]
)

transactions = transactions.drop_duplicates()

transactions = transactions[
    transactions["amount_inr"] > 0
]

transactions["transaction_type"] = (
    transactions["transaction_type"]
    .str.strip()
    .str.title()
)

valid_kyc = ["Verified", "Pending"]

transactions = transactions[
    transactions["kyc_status"].isin(valid_kyc)
]

print("Cleaned Shape:", transactions.shape)
```

```python
print("Missing Values:")
print(transactions.isnull().sum())

print("\nDuplicate Rows:")
print(transactions.duplicated().sum())

print("\nAmount <= 0:")
print((transactions["amount_inr"] <= 0).sum())

print("\nKYC Values:")
print(transactions["kyc_status"].unique())
```

```python
transactions.to_csv(
    "data/processed/08_investor_transactions_cleaned.csv",
    index=False
)

print("Investor Transactions Saved!")
```

```python
print("Shape:", performance.shape)

print("\nMissing Values:")
print(performance.isnull().sum())

print("\nDuplicate Rows:")
print(performance.duplicated().sum())

print("\nExpense Ratio < 0.1:")
print((performance["expense_ratio_pct"] < 0.1).sum())

print("\nExpense Ratio > 2.5:")
print((performance["expense_ratio_pct"] > 2.5).sum())
```

```python
performance = performance.drop_duplicates()

performance = performance[
    (performance["expense_ratio_pct"] >= 0.1)
    &
    (performance["expense_ratio_pct"] <= 2.5)
]

performance["risk_grade"] = (
    performance["risk_grade"]
    .astype(str)
    .str.strip()
)

print("Cleaned Shape:", performance.shape)
```

```python
print("Duplicate Rows:")
print(performance.duplicated().sum())

print("\nExpense Ratio < 0.1:")
print((performance["expense_ratio_pct"] < 0.1).sum())

print("\nExpense Ratio > 2.5:")
print((performance["expense_ratio_pct"] > 2.5).sum())
```

```python
performance.to_csv(
    "data/processed/07_scheme_performance_cleaned.csv",
    index=False
)

print("Scheme Performance Saved!")
```

```python
import pandas as pd
import os

os.makedirs("data/processed", exist_ok=True)

files = [
    "01_fund_master.csv",
    "03_aum_by_fund_house.csv",
    "04_monthly_sip_inflows.csv",
    "05_category_inflows.csv",
    "06_industry_folio_count.csv",
    "09_portfolio_holdings.csv",
    "10_benchmark_indices.csv"
]

for file in files:

    df = pd.read_csv(f"data/raw/{file}")

    df = df.drop_duplicates()

    df.to_csv(
        f"data/processed/{file.replace('.csv','_cleaned.csv')}",
        index=False
    )

    print(file, "saved")
```

```python
import sqlite3

conn = sqlite3.connect("bluestock_mf.db")

print("Database Created Successfully!")
```

```python
import pandas as pd

fund_master = pd.read_csv("data/processed/01_fund_master_cleaned.csv")

nav = pd.read_csv("data/processed/02_nav_history_cleaned.csv")

transactions = pd.read_csv("data/processed/08_investor_transactions_cleaned.csv")

performance = pd.read_csv("data/processed/07_scheme_performance_cleaned.csv")

aum = pd.read_csv("data/processed/03_aum_by_fund_house_cleaned.csv")

print("All files loaded successfully!")
```

```python
print(fund_master.shape)
print(nav.shape)
print(transactions.shape)
print(performance.shape)
print(aum.shape)
```

```python
fund_master.to_sql(
    "dim_fund",
    conn,
    if_exists="replace",
    index=False
)

nav.to_sql(
    "fact_nav",
    conn,
    if_exists="replace",
    index=False
)

transactions.to_sql(
    "fact_transactions",
    conn,
    if_exists="replace",
    index=False
)

performance.to_sql(
    "fact_performance",
    conn,
    if_exists="replace",
    index=False
)

aum.to_sql(
    "fact_aum",
    conn,
    if_exists="replace",
    index=False
)

print("All tables loaded successfully!")
```

```python
cursor = conn.cursor()

tables = [
    "dim_fund",
    "fact_nav",
    "fact_transactions",
    "fact_performance",
    "fact_aum"
]

for table in tables:

    count = cursor.execute(
        f"SELECT COUNT(*) FROM {table}"
    ).fetchone()[0]

    print(table, ":", count)
```

```python
query = """
SELECT scheme_name, aum_crore
FROM fact_performance
ORDER BY aum_crore DESC
LIMIT 5;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT amfi_code,
ROUND(AVG(nav),2) AS avg_nav
FROM fact_nav
GROUP BY amfi_code
ORDER BY avg_nav DESC;
"""

pd.read_sql(query, conn).head()
```

```python
query = """
SELECT state,
COUNT(*) AS total_transactions
FROM fact_transactions
GROUP BY state
ORDER BY total_transactions DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT transaction_type,
COUNT(*) AS total
FROM fact_transactions
GROUP BY transaction_type;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT scheme_name,
expense_ratio_pct
FROM fact_performance
WHERE expense_ratio_pct < 1;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT category,
ROUND(AVG(return_1yr_pct),2) AS avg_return
FROM fact_performance
GROUP BY category;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT state,
ROUND(SUM(amount_inr),2) AS total_amount
FROM fact_transactions
GROUP BY state
ORDER BY total_amount DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT scheme_name,
sharpe_ratio
FROM fact_performance
ORDER BY sharpe_ratio DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT category,
COUNT(*) AS total_funds
FROM dim_fund
GROUP BY category;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT risk_grade,
COUNT(*) AS total_funds
FROM fact_performance
GROUP BY risk_grade;
"""

pd.read_sql(query, conn)
```
